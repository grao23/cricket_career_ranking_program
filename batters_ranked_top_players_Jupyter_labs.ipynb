```python
batters = pd.read_csv("batters_test.csv")

batters.dropna()
batters.head()
```

```python
#batting = []

#def batting_grade(batting): 
    
# Calculate grade based on runs
for i in range(len(batters['Runs'])):
    Runs = batters['Runs'][i]
    avg = batters['AVG'][i]
    strike_rate = batters['Strike Rate'][i]

    # Calculate grade based on runs
    if Runs >= 10000: 
        batters['Runs'][i] = 30 * 1
    elif 9000 <= Runs < 10000: 
        batters['Runs'][i] = 30 * 0.92
    elif 8000 <= Runs < 9000: 
        batters['Runs'][i] = 30 * 0.87
    elif 7000 <= Runs < 8000: 
        batters['Runs'][i] = 30 * 0.79
    elif 6000 <= Runs < 7000: 
        batters['Runs'][i] = 30 * 0.68
    elif 5000 <= Runs < 6000: 
        batters['Runs'][i] = 30 * 0.60
    elif 4000 <= Runs < 5000: 
        batters['Runs'][i] = 30 * 0.53
    elif 3000 <= Runs < 4000:
        batters['Runs'][i] = 30 * 0.40
    elif 2000 <= Runs < 3000: 
        batters['Runs'][i] = 30 * 0.27
    elif 1000 <= Runs < 2000: 
        batters['Runs'][i] = 30 * 0.15
    else: 
        batters['Runs'][i] = 30 * 0.05

    # Calculate grade based on strike rate
    if strike_rate >= 90.00: 
        batters['Strike Rate'][i] = 20 * 1
    elif 80.00 <= strike_rate < 90.00: 
        batters['Strike Rate'][i] = 20 * 0.92
    elif 70.00 <= strike_rate < 80.00: 
        batters['Strike Rate'][i] = 20 * 0.78
    elif 65.00 <= strike_rate < 70.00: 
        batters['Strike Rate'][i] = 20 * 0.70
    elif 60.00 <= strike_rate < 65.00: 
        batters['Strike Rate'][i] = 20 * 0.65
    elif 55.00 <= strike_rate < 60.00: 
        batters['Strike Rate'][i] = 20 * 0.60
    elif 50.00 <= strike_rate < 55.00: 
        batters['Strike Rate'][i] = 20 * 0.53
    elif 45.00 <= strike_rate < 50.00: 
        batters['Strike Rate'][i] = 20 * 0.45
    elif 40.00 <= strike_rate < 45.00: 
        batters['Strike Rate'][i] = 20 * 0.32
    elif 30.00 <= strike_rate < 40.00: 
        batters['Strike Rate'][i] = 20 * 0.20
    else: 
        batters['Strike Rate'][i] = 20 * 0.10

    # Calculate grade based on average
    if avg >= 50.00: 
        batters['AVG'][i] = 35 * 1
    elif 45.00 <= avg < 50.00: 
        batters['AVG'][i] = 35 * 0.90
    elif 40.00 <= avg < 45.00: 
        batters['AVG'][i] = 35 * 0.78
    elif 35.00 <= avg < 40.00: 
        batters['AVG'][i] = 35 * 0.61
    elif 30.00 <= avg < 35.00: 
        batters['AVG'][i] = 35 * 0.48
    elif 25.00 <= avg < 30.00: 
        batters['AVG'][i] = 35 * 0.28
    elif 20.00 <= avg < 25.00: 
        batters['AVG'][i] = 35 * 0.15
    else: 
        batters['AVG'][i] = 35 * 0.05

# Calculate the batter grade
batters_grade = (batters['Runs'] + batters['AVG'] + batters['Strike Rate'] - (batters['Ducks'] * 0.07) + (batters['50s'] * 0.05) + (batters['100s'] * 0.10) + (batters['200s'] * 0.15))

# Add the batter grade to the DataFrame
batters["batters_grade"] = batters_grade
```
